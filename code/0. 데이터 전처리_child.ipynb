# 5년치 병합

- 나이대: 0세~19세


```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('child_2000.csv',encoding='cp949',header=10)
df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed')]
df1['year']=2000
df1.head()
```

```python
df2=pd.read_csv('child_2001.csv',encoding='cp949',header=10)
df2 = df2.loc[:, ~df2.columns.str.contains('^Unnamed')]
df2['year']=2001
df2.head()
```

```python
df3=pd.read_csv('child_2002.csv',encoding='cp949',header=10)
df3 = df3.loc[:, ~df3.columns.str.contains('^Unnamed')]
df3['year']=2002
df3.head()
```

```python
df4=pd.read_csv('child_2003.csv',encoding='cp949',header=10)
df4 = df4.loc[:, ~df4.columns.str.contains('^Unnamed')]
df4['year']=2003
df4.head()
```

```python
df5=pd.read_csv('child_2004.csv',encoding='cp949',header=10)
df5 = df5.loc[:, ~df5.columns.str.contains('^Unnamed')]
df5['year']=2004
df5.head()
```

```python
df_final=pd.concat([df1,df2,df3,df4,df5])
df_final.head()
```

```python
df_final.to_csv('booklist_child_00_04.csv',encoding='cp949')
```
